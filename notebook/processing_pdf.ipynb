```python
import requests
import PyPDF2

def download_pdf(url, save_path='downloaded.pdf'):
    response = requests.get(url)
    with open(save_path, 'wb') as pdf_file:
        pdf_file.write(response.content)

def extract_text_from_pdf(pdf_path):
    with open(pdf_path, 'rb') as pdf_file:
        pdf_reader = extract_text(pdf_file)
    
    return pdf_reader

def main():
    pdf_url = "http://eldata11.topica.edu.vn/HocLieu/TGL101/Giao%20trinh/10_TGL101_Bai7_v1.0014103225.pdf"
    
    # Tải và lưu file PDF từ internet
    download_pdf(pdf_url)

    # Trích xuất văn bản từ file PDF
    extracted_text = extract_text_from_pdf('downloaded.pdf')

    # In ra văn bản trích xuất
    print(extracted_text)

if __name__ == "__main__":
    main()
```
